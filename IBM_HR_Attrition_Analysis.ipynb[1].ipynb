```python
#Attrition rate: The attrition rate measures the percentage of employees who leave then
#company in a given period of time. It is usually calculated within a year and is
#expressed as a percentage of the total number of employees.

#Attrition Rate=( 'Total Employees'/'Number of Attrition = Yes')*100
# Step 1: Total number of employees
total_employees = len(ibm_dataset)

# Step 2: Number of employees who left (Attrition = 'Yes')
employees_left = ibm_dataset[ibm_dataset['Attrition'] == 'Yes'].shape[0]

# Step 3: Calculate attrition rate
attrition_rate = (employees_left / total_employees) * 100

# Step 4: Print the result
print(f"Total Employees: {total_employees}")
print(f"Employees Left: {employees_left}")  
print(f"Attrition Rate: {attrition_rate:.2f}%") 
#f"" stands for formatted string literal — or f-string for short.
#It allows you to insert variables directly inside a string.
#It is used to format numbers, especially floating-point numbers, to 2 decimal places. 

```

```python
#Plotting Attrition Rate
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate value counts as percentages
attrition_percent = ibm_dataset['Attrition'].value_counts(normalize=True) * 100

# Plot bar chart
sns.barplot(x=attrition_percent.index, y=attrition_percent.values, palette='Set2')

plt.title('Distribution of Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

```python
#Average tenure: The average tenure measures the average number of years an 
#employee stays with the company before leaving. 
#It can provide insight into workforce stability and employee satisfaction within the organization
# Average tenure of employees
avg_tenure = ibm_dataset['YearsAtCompany'].mean()

print(f"Average Employee Tenure: {avg_tenure:.2f} years")
#it can be concluded that many employees feel comfortable and have been with the company for a long time
```

```python
#Employee Demographics Analysis
#1️⃣ Gender Distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=ibm_dataset, x='Gender', palette='pastel')
plt.title('Gender Distribution')
plt.show()
ibm_dataset['Gender'].value_counts()


#2️⃣ Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(ibm_dataset['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

#3️⃣ Education Level
# Education mapping (optional for labels)
education_map = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

ibm_dataset['EducationLevel'] = ibm_dataset['Education'].map(education_map)

sns.countplot(data=ibm_dataset, x='EducationLevel', palette='muted')
plt.title('Education Level of Employees')
plt.xticks(rotation=45)
plt.show()

#4️⃣ Department-wise Distribution
sns.countplot(data=ibm_dataset, x='Department', palette='Set2')
plt.title('Employees per Department')
plt.xticks(rotation=20)
plt.show()

#5️⃣ Job Role Breakdown
plt.figure(figsize=(10,6))
sns.countplot(data=ibm_dataset, y='JobRole', palette='coolwarm')
plt.title('Employee Count by Job Role')
plt.show()

#Compare with Attrition
sns.countplot(data=ibm_dataset, x='Gender', hue='Attrition', palette='Set1')

#conclusion
#Age: Most of the company's employees are in the 30-35 age group.This indicates that the company has many employees who are at a productive experienced age.
#2. Gender: The majority of employees at this company are male. There are significantly more male employees than female employees.
#3. Department: Most of the company's employees are concentrated in the research and development department. This indicates that the company is
#heavily focused on product or service research and development activities




```
